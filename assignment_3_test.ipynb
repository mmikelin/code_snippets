```python
# prompt: load the language model Qwen2-0.5B-Instruct and use it tokenizer

!pip install transformers

from transformers import AutoTokenizer, AutoModelForCausalLM

#load the model, you can see the name in huggingface
model_name =  "Qwen/Qwen2-0.5B-Instruct"  #microsoft/Phi-3-mini-128k-instruct", "Qwen/Qwen2-0.5B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Example usage:
text = "Hello, can you answer my question? Can you introduce about yourself?"
input_ids = tokenizer(text, return_tensors="pt").input_ids
outputs = model.generate(input_ids, max_length=50)


generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
# You can then process the outputs further, e.g., decode them to get text.
```

```python

#You can also use chat template to transform the input text's format
#When using GPT-2, you shouldn't apply chate template.
chat = [
  {"role": "system", "content": "You are a helpful assistant."},
  {"role": "user", "content": "Hello, what's your name?"},
]
text = tokenizer.apply_chat_template(chat, tokenize=False)
text
```

```python

input_ids = tokenizer(text, return_tensors="pt").input_ids
outputs = model.generate(input_ids, max_length=500)

generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
```
